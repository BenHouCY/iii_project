```python
autt=[]
aut = list(data['author'])
for i in aut :
    autt.append(i)#.replace('  ',' ').replace('  ',' ').replace('  ',' '))
```

```python
yORn = list(data['only author or not'])
tup = zip(aut,yORn)
```

```python
# for i in tup :
#     print(i)
```

```python
from collections import Counter
cnt = Counter()

for i in tup:
    author=i[0]
    print(author)
    yORn=i[1]
    if author in cnt:
        if yORn in 'Y':
            cnt[author] += 2
        else:
            cnt[author] += 1
            
    else:
        if yORn in 'Y':
            cnt[author] = 2
        else:
            cnt[author] = 1
            
```

```python
#cnt.most_common(100)
```

```python
rankk=[]
for i in cnt.most_common():
    j=i[0].replace('  ',' ')
    rankk.append((j,i[1]))
print(rankk)
```

```python
# from collections import Counter
cntt = Counter()
for i in rankk:
    author=i[0]
    yORn=i[1]
    if author in cnt:
        cntt[author] += i[1]
            
    else:
        cntt[author] = i[1]
```

```python
rankkk=cntt.most_common()
```
