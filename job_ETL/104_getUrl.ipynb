```python
import requests

def getPages(i):
    res = requests.get("https://www.104.com.tw/jobbank/joblist/joblist.cfm?jobsource=n104bank1&ro=0&jobcat=2007000000&order=2&asc=0&page={}".format(i))
    return res
```

```python
from lxml import etree

def getUrl(res):
    
    selector = etree.HTML(res.text)
    urlList=[]
    j=1
    while True:
        links = selector.xpath('//*[@id="jl_page{}"]/div[{}]/ul[1]/li[3]/div/a/@href'.format(i,j))
        j += 1
        if bool(links) == False:
            break
        urlList.append(*links)
    return urlList
```

```python
urlListBig=[]
for i in range(1,151,1):
    res=getPages(i)
    #print(i)
    urlList=getUrl(res)
    #print(urlList)
    urlListBig.append(urlList)
    print(urlListBig)
```
