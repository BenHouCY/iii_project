```python
import requests

def getPages(i):
    res = requests.get("https://www.518.com.tw/job-index-P-{}.html?i=1&am=1&ab=2032001,2032002,&xq=1".format(i))
    return res
```

```python
from lxml import etree


def getUrl(res):
    
    selector = etree.HTML(res.text)
    urlList=[]
    j=1
    while True:
        links = selector.xpath('//*[@id="listContent"]/ul[{}]/li[2]/a/@href'.format(j))
        j += 1
        if bool(links) == False:
            break
        urlList.append(*links)
    return urlList
```

```python
urlListBig=[]
for i in range(1,56,1):
    res=getPages(i)
    #print(i)
    urlList=getUrl(res)
    #print(urlList)
    for url in urlList:
        urlListBig.append(url)
print(*urlListBig)
```
